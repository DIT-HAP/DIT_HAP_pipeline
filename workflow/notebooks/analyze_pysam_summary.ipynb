```python
summary.query("R1_MAPQ > 1 and R1_NCIGAR==1 and R1_NM <=3 and R1_SA.isna() and R1_XA.isna() and Is_Proper_Pair == 'Yes'")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Filter for rows where both R1 and R2 are properly paired
properly_paired = summary[summary['Is_Proper_Pair'] == 'Yes']

# Select numeric columns for log transformation and plotting
numeric_cols = ['R1_MAPQ', 'R1_LEN', 'R1_NCIGAR', 'R2_MAPQ', 'R2_LEN', 'R2_NCIGAR']

# Create subplots for histograms
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    # Get non-null values
    values = properly_paired[col].dropna()
    
    # Apply log transformation (add 1 to handle zeros)
    log_values = np.log10(values + 1)
    
    # Create histogram
    axes[i].hist(log_values, bins=50, alpha=0.7, edgecolor='black')
    axes[i].set_title(f'Log10({col} + 1) Distribution')
    axes[i].set_xlabel(f'Log10({col} + 1)')
    axes[i].set_ylabel('Frequency')
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

properly_paired
```
