# 1. Import

```python
import pandas as pd
import numpy as np
import altair as alt
```

# 2. Load data

```python
data = pd.read_excel("../../../resources/organized_FACS_depletion_curve.xlsx", sheet_name="new_DIT_HAP", index_col=[0,1,2,3])
```

# 3. Plot

```python
transformed_data = data.rename_axis("Timepoint", axis=1).stack().unstack("Feature").reset_index()
```

```python
alt.Chart(transformed_data).mark_line(point=True).encode(
    x='Generations:Q',
    y='LFC:Q',
    color='Strain_alias:N'
).facet(
    'Gene_name:N',
    columns=5
)
```
