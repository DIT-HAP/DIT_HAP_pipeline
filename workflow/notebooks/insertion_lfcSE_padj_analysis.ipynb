```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.Load data

```python
insertion_statistics = pd.read_csv("/data/c/yangyusheng_optimized/DIT_HAP_pipeline/results/test_project/15_insertion_level_depletion_analysis/insertions_LFC.csv", index_col=[0,1,2,3], header=[0,1])

fitting_results = pd.read_csv("/data/c/yangyusheng_optimized/DIT_HAP_pipeline/results/test_project/16_insertion_level_curve_fitting/insertions_LFC_fitted.csv")
```

# 2. Plot the correlation between lfcSE and padj

```python
timepoints = insertion_statistics.columns.get_level_values(0).unique().tolist()

fig, ax = plt.subplots(1, len(timepoints), figsize=(5*len(timepoints), 5), sharex=True, sharey=True)

for i, timepoint in enumerate(timepoints):
    timepoint_data = insertion_statistics[timepoint]
    timepoint_data = timepoint_data.query("padj < 1")
    lfcSE = timepoint_data['lfcSE']
    padj = timepoint_data['padj']
    ax[i].scatter(lfcSE, padj, facecolor='none', edgecolor='grey', alpha=0.5)
    ax[i].set_title(timepoint)
    ax[i].set_xlabel('lfcSE')
    ax[i].set_ylabel('padj')

plt.show()
plt.close()
```

```python
timepoints = insertion_statistics.columns.get_level_values(0).unique().tolist()

fig, ax = plt.subplots(1, len(timepoints), figsize=(5*len(timepoints), 5), sharex=True)

for i, timepoint in enumerate(timepoints):
    timepoint_data = insertion_statistics[timepoint].query("padj < 1")
    lfcSE = timepoint_data['lfcSE'].fillna(0)
    ax[i].hist(lfcSE, bins=100)

plt.show()
plt.close()
```
